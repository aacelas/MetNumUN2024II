```
from sympy import *
x1, x2, x3, y1, y2, y3, z1, z2, z3 = symbols("x1 x2 x3 y1 y2 y3 z1 z2 z3", real=True)
x = x1 + I*x2 + I*x3
y = y1 + I*y2 + I*y3
z = z1 + I*z2 + I*z3

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
